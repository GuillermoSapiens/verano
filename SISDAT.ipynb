```julia
using SACFiles
using PyPlot
```

```julia
dataE=readsac("/home/taller/Desktop/verano/dataVerano/SV.ST03L..EHE.SAC")
dataN=readsac("/home/taller/Desktop/verano/dataVerano/SV.ST03L..EHN.SAC")
dataZ=readsac("/home/taller/Desktop/verano/dataVerano/SV.ST03L..EHZ.SAC");
```

```julia
DaE=dataE.data
DaN=dataN.data
DaZ=dataZ.data;
```

```julia
tE=(collect(0:1:length(DaE)-1))*dataE.hdr.delta
tN=(collect(0:1:length(DaN)-1))*dataN.hdr.delta
tZ=(collect(0:1:length(DaZ)-1))*dataZ.hdr.delta;
```

```julia
plot(tE,DaE)
```

```julia
plot(tN,DaN)
```

```julia
plot(tZ,DaZ)
```
